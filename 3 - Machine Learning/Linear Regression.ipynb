## Linear Regression
- i - Variable i. i = {1,2,3,...M}
- $x_i$ - Name of independent variable $i$.
- $N$ - Number of data points/rows.
- $y$ - Name of target/dependent variable.
- $X$ - Independant variable data. (N x M)
- $Y$ - Dependent variable data. (N x 1)
- $w_i$ linear coefficient relating $x_i$ to $y$. $w$: (M x 1)
- $\epsilon$ - Normally distributed error of model to Y. (N x 1) \

$Y = Xw + \epsilon$ \
$\epsilon^2 = (Y - Xw)^2 $

Optimisation Model:
Minimise{ $\epsilon^2 = (Y - Xw)^2 $ }
